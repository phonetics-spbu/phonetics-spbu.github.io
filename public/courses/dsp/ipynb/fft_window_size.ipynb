```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from scipy.fft import fft, fftfreq, rfft, rfftfreq
from scipy.io import wavfile
```

```python
fs, data = wavfile.read(r"period.wav")
```

```python
plt.plot(data)
```

```python
freqs = rfftfreq(len(data), 1 / fs)
spectrum = np.abs(rfft(data))
```

```python
print(len(data), len(freqs))
```

```python
freqs[-1]
```

```python
freqs[1] - freqs[0]
```

```python
plt.plot(freqs, spectrum)
plt.xlim(0, 3000)
```

```python
period_count = 5

new_period = list(data) * 5
```

```python
plt.plot(new_period)
```

```python
new_freqs = rfftfreq(len(new_period), 1 / fs)
new_spectrum = np.abs(rfft(new_period))

print(len(new_period), len(new_freqs))
```

```python
plt.plot(new_freqs, new_spectrum)
plt.xlim(0, 3000)
```

```python
period_count = int(fs // len(data))
period_count
```

```python
new_period = list(data) * period_count
```

```python
new_freqs = rfftfreq(len(new_period), 1 / fs)
new_spectrum = np.abs(rfft(new_period))

print(len(new_period), len(new_freqs))
```

```python
plt.plot(new_freqs, new_spectrum)
plt.xlim(0, 3000)
```

```python
fs, data_fem = wavfile.read(r"period_fem.wav")
```
