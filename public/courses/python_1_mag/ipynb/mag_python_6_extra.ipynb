Задача 1. Напишите программу, которая проверяет, являются ли две строки анаграммами (содержат одинаковые символы в разном порядке без учета регистра и пробелов). Примеры анаграмм: полусмерть-пульсометр, апельсин-спаниель, ревность-верность, равновесие-своенравие, артшкола-шарлотка, аорист-сирота.

```python
word1 = list(input().lower())
word2 = list(input().lower())

print(sorted(word1) == sorted(word2))
```

Задача 2. Напишите программу, которая проверяет, является ли слово изограммой (все буквы уникальны, без учета регистра).

```python
word = input().lower()

print(len(set(word)) == len(word))
```

Задача 4. Объедините два словаря, суммируя значения для одинаковых ключей. Например, для словарей

dict1 = {'a': 1, 'b': 2}

dict2 = {'b': 3, 'c': 4}

результатом будет {'a': 1, 'b': 5, 'c': 4}

```python
dict1 = {'a': 1, 'b': 2}
dict2 = {'b': 3, 'c': 4}

for key, value in dict2.items():
  if key in dict1.keys():
    dict1[key] = dict1[key] + value
  else:
    dict1[key] = value

print(dict1)
```

Задача 5. Напишите программу, которая принимает две строки одинаковой длины и возвращает количество позиций, в которых символы различаются (расстояние Хэмминга).



```python
word1 = list(input().lower())
word2 = list(input().lower())

hamming_distance = 0

for i, letter in enumerate(word1):
  if letter != word2[i]:
    hamming_distance += 1

print(hamming_distance)
```
