Задача 1. Напишите программу, которая проверяет, являются ли две строки анаграммами (содержат одинаковые символы в разном порядке без учета регистра и пробелов). Примеры анаграмм: полусмерть-пульсомер, апельсин-спаниель, ревность-верность, равновесие-своенравие, артшкола-шарлотка, аорист-сирота.

Задача 2. Напишите программу, которая проверяет, является ли слово изограммой (все буквы уникальны, без учета регистра).

Задача 3. Считайте две строки. Найдите символы, которые есть в первой строке, но отсутствуют во второй.

Задача 4. Объедините два словаря, суммируя значения для одинаковых ключей. Например, для словарей

dict1 = {'a': 1, 'b': 2}

dict2 = {'b': 3, 'c': 4}

результатом будет {'a': 1, 'b': 5, 'c': 4}

Задача 5. Напишите программу, которая принимает две строки одинаковой длины и возвращает количество позиций, в которых символы различаются (расстояние Хэмминга).
