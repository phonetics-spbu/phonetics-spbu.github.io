1. Генератор пророчеств. Напишите функцию, генерирующую пророчества по шаблону:

```
templates = [
  "В {день} ты встретишь {существо}.",
  "Бойся {цвет} {предмета} в {место}.",
  "Твое счастье в {действие} под {небесное тело}."
]
```
Функция должна возвращать предсказание по случайному шаблону, возможные варианты мест/существ/действий и т.д. должны быть оформлены в словарь.

````python
import random

dict_entities = {
    "days": ["понедельник", "Новый Год", "пятницу 13"],
    "creat": ["домового", "воробья", "Филлипа Киркорова"],
    "colors": ["черного", "белого", "лазурный"],
    "items": ["креста", "трамвая", "экзамена"],
    "places": ["болоте", "пышечной", "университете"],
    "actions": ["падении", "полете", "молчании"],
    "stars": ["Луной", "Бетельгейзе", "ретроградным Меркурием"]
}

def get_prophecy():
  print(f"В {random.choice(dict_entities["days"])} ты встретишь {random.choice(dict_entities["creat"])}. Бойся {random.choice(dict_entities["colors"])} {random.choice(dict_entities["items"])} в {random.choice(dict_entities["places"])}. Твое счастье в {random.choice(dict_entities["actions"])} под {random.choice(dict_entities["stars"])}.")
````

````python
get_prophecy()
````
