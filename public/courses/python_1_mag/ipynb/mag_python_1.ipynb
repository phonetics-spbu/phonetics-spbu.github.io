Арифметические операции: +, -, \*, /, \** <br/>

Общие правила определения приоритетов операций такие:<br/>

Сначала выполняются операции возведения в степень справа налево, то есть 2 ** 3 ** 2 — это 2 ** (3 ** 2)=512
.<br/>
Затем выполняются унарные минусы (отрицания).<br/>
Далее выполняются умножения и деления слева направо. Операции умножения и деления имеют одинаковый приоритет.<br/>
И последними выполняются сложения и вычитания слева направо. Операции сложения и вычитания имеют одинаковый приоритет.

```python
print(2**3)
```

Типы данных

Если мы присваиваем переменной a значение, равное 5, то переменная будет иметь значение типа int, который используется для целых чисел. При умножении, вычитании и сложении чисел этого типа мы получаем числа того же типа.

Мы уже сталкивались с тем, что при выполнении операции деления результатом будет дробное число. Если переменной b присвоить значение 10 / 2, при выводе на экран результата увидим число 5.0, которое равно числу 5, но имеет другой тип. Переменная b имеет значение типа float. Это специальный тип данных для представления действительных чисел, и при умножении, вычитании и сложении чисел этого типа мы получаем числа того же типа.

Ещё один тип данных используется для работы со строками. Если мы присваиваем переменной с значение, равное 'Hello', то эта переменная будет иметь значение типа str. Строки тоже можно складывать. Однако операция сложения для целых чисел и для строк работает по-разному: для чисел это сложение, а для строк — конкатенация.



Основные операции над строками:

A + B — конкатенация, то есть соединение строк путём размещения строки B сразу после строки A;

A * n — повторение строки n
 раз, значение n
 должно быть целого типа.

Ввод данных: функция input() <br/>

Для считывания строки с клавиатуры используется функция input(), которая возвращает значение считанной строки. Например, можно считать значения двух строк следующим образом:

b = input()

```python
a = int(input())
print(a * 2)
```

```python
n = int(input("Введите количество ручек "))
price = int(input())

string = f"{n} ручек стоят {n*price} рублей"
print(string)
```

Задание. Напишите программу, вычисляющую индекс массы тела (ИМТ) и выводящую его на печать. На вход программа принимает рост человека в сантиметрах и вес в килограммах. ИМТ рассчитывается как отношение массы тела в килограммах к квадрату роста в метрах.

```python
# ваш код
h = int(input("Введите ваш рост в см: "))
w = int(input("Введите ваш вес в кг: "))

imt = w / (h / 100) ** 2
print("Ваш ИМТ: ", imt)
```

Задача. Следующее и предыдущее<br/>
Напишите программу, которая считывает целое число и выводит текст, аналогичный приведённому в примере. Пробелы, знаки препинания, заглавные и строчные буквы важны!

Входные данные:

Вводится целое число, по модулю не превосходящее 10000

Выходные данные:

Выведите сначала фразу «The next number for the number», затем введённое число, затем слово «is», окружённое пробелами, затем формулу для следующего за введённым числа, наконец, знак точки без пробела. Аналогично в следующей строке для предыдущего числа.

Примечание

Вывод необходимо организовать в две строки, по одному предложению в каждой. При выводе обращайте внимание на пробелы и регистр.

Если в функции print() просто через запятую перечислять аргументы, по умолчанию они будут выводиться через пробел. Чтобы решить эту проблему, используйте конкатенацию (сложение строк), либо f-строку.

Пример:<br/>
Ввод<br/>
179<br/>
Вывод<br/>
The next number for the number 179 is 180.<br/>
The previous number for the number 179 is 178.<br/>

```python
# ваш код
current_number = int(input())

next_number = current_number + 1
prev_number = current_number - 1

print(f"The next number for the number {current_number} is {next_number}.")
print(f"The previous number for the number {current_number} is {prev_number}.")
```
