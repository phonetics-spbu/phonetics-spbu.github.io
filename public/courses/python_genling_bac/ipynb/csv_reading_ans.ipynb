#### Задание 1

По таблице *student_marks.csv* определите количество учеников в классе и составьте алфавитный список.

```python
import csv
```

```python
with open("student_marks.csv", newline="", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    rows = list(reader)

students = sorted({row["фамилия"] + " " + row["имя"] for row in rows})

for i, name in enumerate(students, start=1):
    print(f"{i}. {name}")
```

#### Задание 2

а) для каждого ученика определите средний балл по каждому предмету.

б) определите средневзвешенную оценку по каждому предмету с использованием весов из таблицы *work_types.csv*.

в) составьте таблицу с этими данными и запишите её в отдельный файл .csv. Первый столбец таблицы пусть содержит фамилию и имя, остальные &mdash; средневзвешенную оценку по каждому предмету в алфавитном порядке.

```python
with open("work_types.csv", newline="", encoding="utf-8") as f:
    reader = csv.reader(f)
    work_types = {}
    for k, v in reader:
        work_types[k] = float(v)
```

```python
subjects = sorted({row["предмет"] for row in rows})

new_table = []
new_header = ["имя"] + subjects

for name in students:
    row = {"имя": name}
    for subj in subjects:
        grades = [
            row for row in rows
            if row["фамилия"] + " " + row["имя"] == name
            and row["предмет"] == subj
        ]
        numerator = sum(int(g["оценка"]) * work_types[g["вид работы"]] for g in grades)
        denominator = sum(work_types[g["вид работы"]] for g in grades)
        weighted_avg = numerator / denominator if denominator else 0
        row[subj] = f"{weighted_avg:.2f}"
    new_table.append(row)

with open("avg_grades.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=new_header)
    writer.writeheader()
    writer.writerows(new_table)
```

#### Задание 3

Найдите ученика с максимальным средним баллом по информатике.

```python
print(max(new_table, key=lambda x: x["информатика"])["имя"])
```

#### Задание 4

Файл *messages.csv* содержит данные о сообщениях в некотором чате. Доступна информация: имя пользователя, дата сообщения, количество слов. Найдите среднее, максимальное и минимальное количество слов в сообщениях каждого пользователя. Найдите пользователя с самым длинным сообщением, отправленным после 18:00.

```python
with open("messages.csv", newline="", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    rows = list(reader)

for row in rows:
    date, time = row["datetime"].split("T")
    h, m, s = [int(i) for i in time.split(":")]
    row["date"] = date
    row.update(dict(h=h, m=m, s=s))
    row["words"] = int(row["words"])

users = sorted(set(row["username"] for row in rows))

for usr in users:
    usr_msg = [row for row in rows if row["username"] == usr]
    max_len = max([row["words"] for row in usr_msg])
    min_len = min([row["words"] for row in usr_msg])
    avg_len = sum([row["words"] for row in usr_msg]) / len(usr_msg)

    print(usr, max_len, min_len, avg_len)

print("---")

longest_msg_after_18 = {}

for usr in users:
    msgs = [row["words"] for row in rows if row["username"] == usr and row["h"] >= 18]
    longest_msg_after_18[usr] = max(msgs) if msgs else 0

print(max(longest_msg_after_18, key=longest_msg_after_18.get))
```
