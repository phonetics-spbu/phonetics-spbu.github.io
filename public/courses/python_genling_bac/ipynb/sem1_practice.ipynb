б) Напишите программу, которая читает текст из файла, проводит необходимую предобработку и записывает результат работы функции из пункта (а) в файл .json.

#### Задание 6

а) Напишите фукнцию, которая принимает на вход строку, содержащую слово на русском языке, и возвращает длину самого длинного сочетания согласных в нём. 

б) Напишите функцию, которая принимает на вход строку, содержащую текст на русском языке, и возвращает список слов из этого текста, отсортированных в порядке возрастания длины самого длинного сочетания согласных в них. Для этого функцию из пункта (а) передайте в параметр `key` функции `sorted()` или метода списков `.sort()`.

в) Протестируйте функции на любом текстовом файле.

#### Задание 7

а) Напишите функцию, которая принимает на вход строку, содержащую транскрипцию для русского языка (обозначения по Щербе) и возвращающую кортеж строк, каждая из которых содержит один звук. Каждый звук может быть обозначен как одним символом, так и двумя. Второй символ обязательно один из следующих:

* ' &mdash; мягкость
* 0 &mdash; ударность гласного
* 8 &mdash; второстепенное ударение


б) Напишите программу, которая читает текст из файла `sibling_sample.txt` в транскрипции и записывает в новый файл .csv таблицу частотности звуков в порядке убывания. В таблице должно быть две колонки: звук и число раз, которое он встретился в тексте. Не забывайте использовать кодировку UTF-8!

в) Напишите программу, которая строит таблицу сочетаемости фонем: по строчкам &mdash; первая фонема в бифонемном сочетании, по столбикам &mdash; вторая, в ячейках &mdash; количество раз, которое встретилось это сочетание. Считайте, что границ между словами нет.
