В библиотеке есть два способа работать с корпусными данными.

1. `opencorpora.load`

Этот метод загружает файл в память целиком. Это удобнее, но может занимать много памяти.

```python
corpus = opencorpora.load("annot.opcorpora.no_ambig.xml")
```

Функция `load()` вернёт нам объект `Corpus`.

```python
corpus.version
```

Иерархия:

Corpus -> Doc -> Paragraph -> Sentence -> Token

В каждом уровне есть поля для всех более нижних уровней (`docs`, `paragraphs`, `sentences`, `tokens`).

2. `CorpusReader`

```python
corpus = opencorpora.CorpusReader("annot.opcorpora.no_ambig.xml")
```

```python
corpus.catalog()[:10]
```

```python
corpus.documents(categories='Тема:ЧасКор:Книги*')[:10]
```

```python
words = corpus.words("2")
```

```python
words[:10]
```
