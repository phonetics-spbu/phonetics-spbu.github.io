Задача 1. Напишите программу, которая принимает строку и возвращает количество гласных букв в ней. (Язык можете выбрать любой, с алфавитной письменностью)

```python
vowels = "ёуеыаоэяию"
string = input()

count = 0
for char in string:
  if char.lower() in vowels:
    count += 1

print(count)
```

Задача 3. Напишите программу, которая проверяет, является ли строка валидным e-mail адресом:
1. содержит @
2. после @ есть хотя бы одна точка
3. Домен длиной не меньше 2 (то, что идет после последней точки, например, com или ru)
4. Не содержит пробелов

```python
email = input()
if "@" in email and "." in email.split("@")[-1] and len(email.split(".")[-1]) > 1 and " " not in email.strip():
  print("email is valid")
else:
  print("not valid")
```

Задача 4. Напишите программу, которая принимает на вход строку с числами и находит в ней "счастливые" (те, у которых сумма цифр четная). Например, на строку "123 456 789 100" программа должна вернуть список [123, 789]

```python
numbers = input().split()

lucky_numbers = []
for number in numbers:
  digits = list(number)
  digits_numeric_sum = sum([int(x) for x in digits])
  if digits_numeric_sum % 2 == 0:
    lucky_numbers.append(number)

print(lucky_numbers)
```

Задача 5. Напишите программу, которая строку в snake_case переводит в строку в camelCase. Например, what_a_year_this_week_had_been будет переведена в whatAYearThisWeekHadBeen

```python
x = input()
word_list = x.split("_")
camelCaseStr = word_list[0]

for word in word_list[1:]:
  camelCaseStr += word.capitalize()

print(camelCaseStr)
```
