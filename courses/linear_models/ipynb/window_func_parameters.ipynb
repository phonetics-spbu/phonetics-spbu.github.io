```python
from scipy import signal
import matplotlib.pyplot as plt
from numpy.fft import fft, fftfreq, fftshift
import numpy as np
```

```python
num_samples = 51
```

```python
rect_w = signal.get_window("boxcar", num_samples)

rect_w_spectrum =  np.abs(fft(rect_w, 2048))
frequencies = np.linspace(-0.5, 0.5, len(rect_w_spectrum))

A = fft(rect_w, 2048)
response = 20 * np.log10(np.abs(fftshift(A)))

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))

ax1.plot(np.abs(rect_w))
ax1.set_title('Boxcar / rectangular / Dirichlet window')
ax1.set_xlabel("Sample")
ax1.set_ylabel("Amplitude")
ax1.grid(True)

ax2.plot(frequencies, fftshift(rect_w_spectrum / abs(rect_w_spectrum).max()))
ax2.set_title('Boxcar window spectrum')
ax2.set_xlabel("Frequencies")
ax2.set_ylabel("Spectral density")
ax2.grid(True)

ax3.plot(frequencies, response)
ax3.set_title('Frequency response of the boxcar window')
ax3.set_xlabel("Normalized magnitude [dB]")
ax3.set_ylabel("Normalized frequency [cycles per sample]")
ax3.grid(True)
```

```python
plt.plot(frequencies, response)
plt.axis([-0.5, 0.5, -120, 0])
plt.title("Frequency response of the boxcar window")
plt.ylabel("Normalized magnitude [dB]")
plt.xlabel("Normalized frequency [cycles per sample]")
plt.xlim(0, 0.15)
plt.ylim(-80, 40)
plt.grid()
```
