Задание 1.

Напишите функцию генерации глухого щелевого звука. Аргументы -  какой генерируем звук (ключи словаря fricatives), длительность в секундах, частота дискретизации.

1. Задайте источник - белый шум нужной длительности
2. Создайте полосовой фильтр (например, Баттерворта 4 порядка). Верхнюю и нижнюю частоту вычислите из значений, данных в словаре.
3. Верните результат фильтрации

```python
def generate_fricative(cons, duration, fs):
  noise = np.random.normal(0, 1, int(duration * fs))

  nyqwist = fs / 2
  low = (fricatives[cons][0] - fricatives[cons][1] / 2) / nyqwist
  high = (fricatives[cons][0] + fricatives[cons][1] / 2) / nyqwist

  b, a = butter(4, [low, high], btype="band")

  return lfilter(b, a, noise)
```

```python
s = generate_fricative("s", 1, fs)
sh = generate_fricative("sh", 1, fs)
f = generate_fricative("f", 1, fs)
```

Задание 2.

Добавьте возможность генерации звонкого щелевого звука. Новые опциональные аргументы - флаг звонкости, отношение амплитуды шума к амплитуде гармонической составляющей, частота основного тона.

1. Сгенерируйте глухой щелевой звук
2. Создайте синусоидальный сигнал с частотой равной ЧОТ и еще пару гармоник с меньшей амплитудой (например, 0.5 и 0.2 от амплитуды первой)
3. Сложите глухой щелевой звук с синусоидой так, чтобы соблюдалась заданная пропорция сигнал-шум, верните результат

```python
def generate_fricative(cons, duration, fs, is_voiced=False, noise_amplitude=0.3, F0=100, make_envelope=True):
  noise = np.random.normal(0, 1, int(duration * fs))

  nyqwist = fs / 2
  low = (fricatives[cons][0] - fricatives[cons][1] / 2) / nyqwist
  high = (fricatives[cons][0] + fricatives[cons][1] / 2) / nyqwist

  b, a = butter(4, [low, high], btype="band")

  result = lfilter(b, a, noise)

  if is_voiced:
    t = np.linspace(0, duration, int(duration * fs))
    voiced_amplitude = 1 - noise_amplitude

    result *= noise_amplitude

    voiced = voiced_amplitude * np.sin(2 * np.pi * f0 * t)
    voiced += 0.5 * voiced_amplitude * np.sin(2 * np.pi * 2 * f0 * t)
    voiced += 0.2 * voiced_amplitude * np.sin(2 * np.pi * 3 * f0 * t)

    result += voiced

  if make_envelope:
    start = int(0.04 * fs)
    end = int(0.05 * fs)

    envelope = np.ones_like(result)
    envelope[:start] = np.linspace(0, 1, start)
    envelope[-end:] = np.linspace(1, 0, end)
    result *= envelope

  return result
```

```python
z = generate_fricative("s", 1, fs, is_voiced=True)
zh = generate_fricative("sh", 1, fs, is_voiced=True)
v = generate_fricative("f", 1, fs, is_voiced=True)
```
