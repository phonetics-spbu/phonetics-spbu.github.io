Folium &mdash; простая библиотека для взаимодействия с картами. В Google Colab установлена по умолчанию, на локальных машинах устанавливается через `pip install folium`.

Документация: https://python-visualization.github.io/folium/latest/

```python
import folium
```

```python
m = folium.Map(location=(59.939095, 30.315868), tiles="Cartodb Positron")
```

В Google Colab вы можете визуализировать карту:

```python
m
```

В обычных скриптах Python вы можете сохранить карту как html и открыть её в браузере:

```python
m.save("mymap.html")
```

Дополнительные параметры:

```python
m = folium.Map(
    location=(59.939095, 30.315868),
    zoom_start=5, # уровень приближения в начале
    max_bounds=True,  # не выходить за пределы
    min_lat=50,
    max_lat=70,
    min_lon=20,
    max_lon=40,
    tiles="Cartodb Positron"
)
m
```

Добавить маркеры:

```python
spbu_buildings = {
    "Филологический факультет": [59.939854, 30.299655],
    "Факультет психологии": [59.945237, 30.298237],
    "Факультет свободных искусств и наук": [59.930402, 30.285069],
}

m = folium.Map(location=(59.939095, 30.315868), zoom_start=13, tiles="Cartodb Positron")

for name, coords in spbu_buildings.items():
    folium.Marker(
        location=coords,
        popup=name,
        color="red"
    ).add_to(m)

m
```

```python
dorms = {
        "Общежитие 5": [59.936713, 30.244886]
}

for name, coords in dorms.items():
    folium.Marker(
        location=coords,
        popup=name,
        icon=folium.Icon(
            color="red",
            icon="asterisk"
        )
    ).add_to(m)

m
```

Цвета маркеров: ['red', 'blue', 'green', 'purple', 'orange', 'darkred',
 'lightred', 'beige', 'darkblue', 'darkgreen', 'cadetblue',

 'gray', 'black', 'lightgray']

```python
m
```

То же самое руками:

```python
import json

with open("export.geojson", encoding="utf-8") as f:
    data = json.load(f)
```

```python
coords = data["features"][0]["geometry"]["coordinates"]

# в формате GeoJSON сначала идёт долгота, потом широта

for i in range(len(coords)):
    for j in range(len(coords[i])):
        coords[i][j] = coords[i][j][::-1]
```

```python
folium.Polygon(
    locations=coords,
    color="blue",
    weight=3,
    fill_color="red",
    fill_opacity=0.2,
    fill=True,
    popup="Василеостровский район"
).add_to(m)

m
```

Нарисовать линию:

```python
points = [
    spbu_buildings["Филологический факультет"],
    [59.938299, 30.294524],
    [59.945288, 30.286157],
    [59.934586, 30.249354],
    [59.932945, 30.244677],
    [59.936430, 30.243097],
    dorms["Общежитие 5"],
]
```

```python
folium.PolyLine(
    locations=points,
    color="#ff00ff",
    weight=5,
    tooltip="от филфака до общаги"
).add_to(m)
m
```
